### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
import xmltodict
dirs = os.listdir( './data' )

# 输出所有文件和文件夹
with open("./data/64_72hr_CH.xml", "r") as file_list:
    doc = dict(xmltodict.parse(file_list.read()))
    
count = 0
for i in doc["cwbopendata"]["dataset"]["locations"]["location"]:
    #print(i["weatherElement"])
    for j in i["weatherElement"]:
        if j["description"] == "溫度":
            count+=1
print(count)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for local in doc["cwbopendata"]["dataset"]["locations"]["location"]:
    for t in local["weatherElement"]:
        if t["description"] == "溫度":
            print((local["locationName"],t["time"][0]["dataTime"],t["time"][0]["elementValue"]["value"]))
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

for local in doc["cwbopendata"]["dataset"]["locations"]["location"]:
    for temp in local["weatherElement"]:
        if temp["description"] == "溫度":
            for k in temp["time"]:
                print(k["dataTime"],k["elementValue"]["value"])
    break
```

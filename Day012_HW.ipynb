### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
#print(soup)
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd

for d in soup.find(class_="part_list_2").find_all('h3'):
    title = d.find_all('a')[-1].text
    date = d.find(class_="date")
    group = d.find_all('em')[-1].text
    news.append({'group':group, 'date':date, 'title':title})
    #print(d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text)
news_df = pd.DataFrame(news).sort_values('group')
#news_df.set_index('group', inplace=True)

print (news_df)
```
